```python
import pandas as pd
import requests

url = 'https://api.covidtracking.com/v1/states/current.json'
response = requests.get(url)

if response.status_code == 200:
    json_data = response.json()
    df = pd.DataFrame(json_data).transpose()
    df.columns = df.loc['state']  # Suponiendo que 'state' contiene las abreviaturas de los estados
    df = df.drop('state')  # Eliminando la fila 'state' ya que se usará para los nombres de las columnas
    print(df.head())
else:
    print(f'Error: {response.status_code}')
```
